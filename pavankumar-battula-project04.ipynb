```python
import pandas as pd
from pathlib import Path

myfiles = [Path(f'/anvil/projects/tdm/data/noaa/{year}.csv') for year in range(1800, 1851)]

for filepath in myfiles:
   
    length = 0
    total = 0

    year = int(filepath.stem)  
    year2= year;
    myDF = pd.read_csv(filepath, header = None, names=["id","date","element_code","value","mflag","qflag","sflag","obstime"]) 

    for index, row in myDF.iterrows():
        if (row['element_code'] == 'PRCP') & ( year2 == year):
            length += 1
            total += row['value']
    avg = total / length
       
    print(f'Year: {year}, Average Precipitation: {avg}')
    
```

```python
year_index = 0


while year_index < len(myfiles):
    filepath = myfiles[year_index]
    length = 0
    total = 0

    year = int(filepath.stem) 
    myDF = pd.read_csv(filepath, header=None, names=["id", "date", "element_code", "value", "mflag", "qflag", "sflag", "obstime"])

    for index, row in myDF.iterrows():
        if row['element_code'] == 'PRCP':
            length += 1
            total += row['value']
    avg = total / length

    if avg < 22:
        print(f'Year: {year}, Average Precipitation: {avg}')
    else:
        break

    year_index += 1
```
